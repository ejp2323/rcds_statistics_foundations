## Problem 1: simulating a coffee grinder
A grinder at coffee house is designed to dispense 200 g of ground coffee per bag. Small random variation in the machine output is common, and the distribution of bag weights can be approximated to a Gaussian:
$$
X \sim \cal{N}(\mu, {\sigma}^2)
$$

with $\mu = 200$ g, $\sigma = 4$ g

To simulate the filling of $n=400$ bags of coffee run the cell below:

```python
rng = np.random.default_rng(0)
n = 400
mu, sigma = 200.0, 4.0
X = rng.normal(mu, sigma, size=n)
```

**Q1**: Plot the weight vs the bag number

```python
#Solution

plt.figure()
plt.plot(X)
plt.xlabel("Bag number")
plt.ylabel("Weight (g)")
plt.title("Coffee bag weights over time")
plt.tight_layout()
plt.show()
```

**Q2**: Plot a histogram of the coffee bag weights

```python
#Solution 

plt.figure()
plt.hist(X, bins=30, density=True, edgecolor="black")
plt.xlabel("Weight (g)")
plt.ylabel("probability density")
plt.title("Coffee bag weights")
plt.tight_layout()
plt.show()
```

**Q3**: Estimate empirically:
- $P(X<195)$
- $P(198<X<202)$

```python
#Solution
p_under = np.mean(X < 195)
p_interval = np.mean((X > 198) & (X < 202))
print("P(X<195) ≈", p_under)
print("P(198<X<202) ≈", p_interval)
```

**Q4**: Standarise the distribution computing $Z = (X-\mu)/\sigma$, and plot the histogram of $Z$

```python
#Solution 
Z = (X - X.mean()) / X.std(ddof=1)
plt.figure()
plt.hist(Z, bins=30, density=True, edgecolor="black")
plt.xlabel("z")
plt.ylabel("Probability density")
plt.title("Standardized weights")
plt.tight_layout()
plt.show()
```

## Problem 2: delivery times

Your food delivery time is skewed: most days are quick, but sometimes it’s really late. A simple model describing the time you need to wait for your food is: 

$$
X \sim \lambda e^{-\lambda}
$$

with $\lambda = 1/30$

This is an an exponential distribution (right-skewed), which is definitely not Gaussian. 

You can simulate $n = 5000$ delivery times with the cell below, in which the scale is defined as $1/\lambda$

```python
rng = np.random.default_rng(1)
scale = 30.0 #scale = mean waiting time 

n = 5000
X = rng.exponential(scale=scale, size=n)
```

**Q1**: After the kth delivery, the mean waiting time can be defined as:
$$
m_{k} = \frac{1}{k} \sum_{i=1}^{k}X_{i}
$$

Calculate the mean waiting time after each delivery and plot it against the delivery number. What do you notice?


```python
#Solution
mean_times = np.cumsum(X) / np.arange(1, n+1)
plt.figure()
plt.plot(mean_times)
plt.axhline(scale, linestyle="--")
plt.xlabel("Number of deliveries observed")
plt.ylabel("Mean waiting time (min)")
plt.title("Average delivery time stabilizes")
plt.tight_layout()
plt.show()
```

**Q2**: Consider $n\in {5, 50, 5000}$. For each value of n:
- Repeat the simulation in **Q1** 1000 times
- Compute the sample mean
- Plot the histogram of those sample means
How does the shape change as $n$ increases?

```python
#Solution 
repeats = 1000
ns = [5, 50, 5000]
for n in ns:
    samples = rng.exponential(scale=scale, size=(repeats, n))
    means = samples.mean(axis=1)

    plt.figure()
    plt.hist(means, bins=40, density=True, edgecolor="black")
    plt.xlabel("Average delivery time (minutes)")
    plt.ylabel("Density")
    plt.title(f"Sampling distribution of the mean (n={n})")
    plt.tight_layout()
    plt.show()
```

## Problem 3: customers in a shop

A bookshop counts how many customers enter each minute during a quiet hour. The distribution of customers approaches a Poissonian:
$$
X \sim Poisson(\lambda)
$$
where $\lambda$ is the average number of arrivals per minute. Let the true rate be $\lambda=3$ customers/min throughout this problem


**Q1**: Simulate arrivals during a large number ($n=60$) of minutes. You will need to get an array of {$N_1, N_2, ... N_{60}$} counts

```python
#Solution 
n = 60
true_rate = 30
sample = rng.poisson(lam=true_rate, size=n)
```

**Q2**: Compute the estimated rate:
$$
\hat{\lambda} = \={N}
$$

```python
#Solution 
lam_hat = sample.mean()
print("Estimated rate: ", lam_hat)
```

**Q3**: $\={N}$ is approximately normal, ie:
$$
\={N} \approx N (\lambda, \frac{\lambda}{n})

$$
Why is this the case?

```python
#Solution
print("Because n is large")
```

**Q4**: The standard error of $\={N}$ is: 
$$
SE(\={N}) = \sqrt{Var(\={N})} = \sqrt\frac{\lambda}{n}
$$
So the approximate 95% CI is: 
$$
\hat{\lambda} \pm 1.96 \sqrt\frac{\lambda}{n}
$$
Compute this interval and check whether it contains the true $\lambda = 3$


```python
#Solution 
se = np.sqrt(lam_hat / n)          # plug-in SE using lam_hat
half = 1.96 * se
low, high = (lam_hat - half, lam_hat + half)
print(f"Interval: ({low, high})")
```

**Q5**: Repeat this experiment 1000 times:
- Simulate $n=60$ min 
- Compute the CI each time
- Record whether it contains $\lambda = 3$
In what fraction of the simulations does the CI contain $\lambda = 3$?

```python
n = 60

def ci95_poisson_clt(sample):
    n = len(sample)
    lam_hat = sample.mean()
    se = np.sqrt(lam_hat / n)          
    half = 1.96 * se
    return lam_hat - half, lam_hat + half, lam_hat

hits=0
for i in range(1000):
    s = rng.poisson(lam=true_rate, size=n)
    lo, hi, _ = ci95_poisson_clt(s)
    hits += (lo <= true_rate <= hi)
cov = hits / repeats

print(f"Estimated coverage ≈ {cov:.5f}")
```
